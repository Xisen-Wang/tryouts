```python
import pandas as pd
import numpy as np
import os
os.getcwd()
os.chdir(r'/Users/wangxiang/Desktop/')
```

```python
df1 = pd.read_excel('Student_Information.xls')
```

```python
df1.head(5)#dataframe 格式
```

```python
df1.loc[df1['入学年']==2016,'入学年']=1111 #修改了column，不加后面的可以单纯筛选
```

```python
df1.head(5)
```

```python
df1.rename(columns={'年级':'grade','姓名':'name','班级':'class','性别':'gender'})#只是看看效果
```

```python
df1.rename(columns={'年级':'grade','姓名':'name','班级':'class','性别':'gender'},inplace=True)#来真的
```

```python
df1.head(5)
```

```python
df1.rename(index={3:3.14},inplace=True)#change rows, 话说就加inplace啦 
```
